```python
import pandas as pd
```

```python
df = pd.read_csv('data_labeled.csv')
df.drop('Unnamed: 0', axis=1, inplace=True)
df.head(10)
```

# 5 Point Analysis w.r.t KMeans Clusters

```python
for i in  df.select_dtypes(exclude='O').drop(['km_labels','ag_labels'],axis=1).columns:
    print('FEATURE NAME-',i,'\n\n5 POINT ANALYSIS OF THE FEATURE-\n',df.groupby('km_labels')[i].describe(),'\n\n\n')
```

```python
for i in  df.select_dtypes(include='O').columns:
    print('FEATURE NAME-',i,'\n\n5 POINT ANALYSIS OF THE FEATURE-\n',df.groupby('km_labels')[i].describe(include='O'),'\n\n\n')
```

# 5 Point Analysis w.r.t Agglomerative Clusters
